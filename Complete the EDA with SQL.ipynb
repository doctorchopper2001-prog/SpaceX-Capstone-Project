```python
import pandas as pd
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")
df.to_sql("SPACEXTBL", con, if_exists='replace', index=False,method="multi")
```

```python
#DROP THE TABLE IF EXISTS

%sql DROP TABLE IF EXISTS SPACEXTABLE;
```

```python
%sql create table SPACEXTABLE as select * from SPACEXTBL where Date is not null
```

```python
%sql SELECT DISTINCT "Launch_Site" FROM SPACEXTABLE;
```

```python
%sql SELECT * FROM SPACEXTABLE WHERE "Launch_Site" LIKE 'CCA%' LIMIT 5;
```

```python
%sql SELECT SUM(CAST("Payload_Mass" AS FLOAT)) AS Total_Payload_Mass_kg FROM SPACEXTABLE WHERE "Customer"='NASA (CRS)';
```

```python
%sql SELECT AVG(CAST("Payload_Mass" AS FLOAT)) AS Avg_Payload_Mass_kg FROM SPACEXTABLE WHERE "Booster_Version"='F9 v1.1';
```

```python
%sql SELECT MIN("Date") AS First_Successful_Ground_Pad_Landing FROM SPACEXTABLE WHERE "Landing_Outcome"='Success (ground pad)';
```

```python
%sql SELECT DISTINCT "Booster_Version" FROM SPACEXTABLE WHERE "Landing_Outcome"='Success (drone ship)' AND CAST("Payload_Mass" AS FLOAT) > 4000 AND CAST("Payload_Mass" AS FLOAT) < 6000;
```

```python
%sql SELECT "Launch_Outcome", COUNT(*) AS Total FROM SPACEXTABLE GROUP BY "Launch_Outcome";
```

```python
%sql SELECT "Booster_Version", "Payload_Mass" FROM SPACEXTABLE WHERE CAST("Payload_Mass" AS FLOAT) = (SELECT MAX(CAST("Payload_Mass" AS FLOAT)) FROM SPACEXTABLE);
```

```python
%sql SELECT substr("Date",6,2) AS Month, "Landing_Outcome", "Booster_Version", "Launch_Site" FROM SPACEXTABLE WHERE "Landing_Outcome"='Failure (drone ship)' AND substr("Date",1,4)='2015';
```
